# Unsupervised machine learning
Unsupervised machine learning is a type of machine learning where the algorithm is given input data without explicit instructions on what to do with it. In other words, the algorithm tries to find patterns, relationships, or structures in the data without labeled responses to guide the learning process. The goal of unsupervised learning is typically to explore the underlying structure of the data or to discover hidden patterns within it.

There are two main types of unsupervised learning techniques:

1. **Clustering:**
   - Clustering algorithms group similar data points together based on some similarity or distance metric. The goal is to identify natural groupings or clusters within the data. Common clustering algorithms include k-means clustering, hierarchical clustering, and DBSCAN.

2. **Dimensionality Reduction:**
   - Dimensionality reduction techniques aim to reduce the number of features or variables in the data while preserving its important characteristics. This is often done to simplify the data and remove redundant information. Principal Component Analysis (PCA) and t-Distributed Stochastic Neighbor Embedding (t-SNE) are examples of dimensionality reduction methods.

**Key Characteristics of Unsupervised Learning:**

1. **No Labeled Output:**
   - Unsupervised learning does not involve labeled output or target values. The algorithm explores the inherent structure of the data without explicit guidance on what to predict.

2. **Exploratory Nature:**
   - The primary purpose of unsupervised learning is often exploratory. It helps in understanding the data, discovering patterns, and gaining insights.

3. **Applications:**
   - Unsupervised learning is widely used in various applications, including data clustering, anomaly detection, feature learning, and topic modeling.

**Common Use Cases:**

- **Clustering Customers:** Identify groups of customers with similar purchasing behavior.
  
- **Anomaly Detection:** Detect unusual patterns or outliers in data.
  
- **Topic Modeling:** Uncover hidden topics in a collection of documents.

- **Dimensionality Reduction:** Reduce the number of features in high-dimensional data.

- **Data Compression:** Represent data in a more compact form while retaining important information.

Unsupervised learning is contrasted with supervised learning, where the algorithm is trained on labeled data with known outputs. Unsupervised learning is particularly valuable in scenarios where labeled data is scarce or expensive to obtain, and the emphasis is on extracting meaningful insights from the raw data itself.

# Clustering
Cluster analysis is a technique used in data mining and machine learning to group similar objects into clusters. K-means clustering is a widely used method for cluster analysis where the aim is to partition a set of objects into K clusters in such a way that the sum of the squared distances between the objects and their assigned cluster mean is minimized.



# K mean clustering

It is a popular clustering algorithm used in unsupervised machine learning. The primary goal of the k-means algorithm is to partition a dataset into a specified number of clusters (k) based on the similarity of data points. It's widely used for tasks like customer segmentation, image compression, and anomaly detection.

## How do we select the "k" value?

Choosing the right value for
k (the number of clusters) is a crucial step in the k-means clustering algorithm. There are several methods to determine the optimal
k, and the choice often depends on the characteristics of your data. Here are some common approaches:

1.  Elbow Method:

The Elbow Method involves running the k-means algorithm for a range of values of
k and plotting the sum of squared distances (inertia or within-cluster sum of squares) for each
k. The "elbow" in the plot represents a point where increasing
k further doesn't significantly decrease the sum of squared distances.

This method uses the concept of WCSS value. WCSS stands for Within Cluster Sum of Squares, which defines the total variations within a cluster.

$
WCSS = ∑ distance(P_i, C_i)^2
$
<br><br>
We compute distance using Eculidean Distance or Manhattan distance.

$
ED = \sqrt(x_2-x_1)^2 + (y_2 -y_1)^2
$
<br><br>
$
MD = |x_2-x_1| + |y_2-y_1|
$

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data
X, y = make_blobs(n_samples=300, centers=4, random_state=42, cluster_std=1.0)

# Elbow Method
inertias = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

# Plotting the Elbow Curve
plt.plot(range(1, 11), inertias, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Sum of Squared Distances (Inertia)')
plt.title('Elbow Method')
plt.show()
````

2.  Silhouette Score:

  The Silhouette Score measures how similar an object is to its own cluster (cohesion) compared to other clusters (separation). A higher silhouette score indicates better-defined clusters.

  In this case, the optimal
k is associated with the highest silhouette score.

````python
from sklearn.metrics import silhouette_score

X, y = make_blobs(n_samples=300, centers=4, random_state=42, cluster_std=1.0)

silhouette_scores = []

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    score = silhouette_score(X, kmeans.labels_)
    silhouette_scores.append(score)

# Plotting Silhouette Scores
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score Method')
plt.show()
````

3. Gap Statistics:

  Gap Statistics compare the within-cluster dispersion of the data to a null reference distribution. The optimal k is the one that maximizes the gap between the data's clustering result and that of a random distribution.

## Random Initialization Trap (K mean ++)

The "random initialization trap" refers to an issue in k-means clustering where the final clusters can be highly dependent on the initial random placement of cluster centroids. Since k-means involves updating the positions of centroids based on the mean of data points assigned to each cluster, starting with different initial centroids can lead to different final cluster assignments.

Here are some key points related to the random initialization trap:

1. **Sensitivity to Initial Centroids:**
   - K-means is sensitive to the initial placement of cluster centroids. If the initial centroids are chosen poorly, the algorithm might converge to a suboptimal solution.

2. **Implications:**
   - Different initializations can result in different clusters and, in extreme cases, can lead to different interpretations of the underlying structure of the data.

3. **Mitigation Techniques:**
   - Several techniques are used to mitigate the impact of random initialization:
     - **Multiple Runs:** Run the k-means algorithm multiple times with different initializations and choose the solution with the lowest inertia.
     - **K-Means++ Initialization:** A smart initialization method that probabilistically chooses initial centroids in a way that makes it less likely to converge to a suboptimal solution.

4. **K-Means++ Initialization:**
   - K-means++ is a modification to the standard k-means algorithm for initializing cluster centroids. It improves the initial selection of centroids by spreading them out across the dataset. This can lead to a more stable convergence.

   ```python
   kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
   ```

To address the random initialization trap, it's common to run k-means with multiple random initializations (e.g., using the `n_init` parameter in scikit-learn) and choose the solution with the lowest inertia. Additionally, using K-Means++ initialization can improve the stability of the algorithm.

## **Limitations**

The k-means algorithm is widely used and effective for many clustering tasks, but it has some limitations and assumptions that can impact its performance in certain scenarios. Here are some of the key limitations of the k-means algorithm:

1. **Sensitivity to Initial Centroids:**
   - K-means is sensitive to the initial placement of cluster centroids. Different initializations can lead to different final clusters. To mitigate this, the algorithm is often run multiple times with different initializations, and the best result is chosen.

2. **Assumption of Spherical Clusters:**
   - K-means assumes that clusters are spherical and equally sized. This means it may not perform well when dealing with clusters of different shapes or sizes.

3. **Assumption of Equal Variance:**
   - K-means assumes that clusters have equal variance. If the variances are significantly different, the algorithm may produce suboptimal results.

4. **Difficulty with Non-Globular Shapes:**
   - K-means struggles with clusters that have non-globular (non-spherical) shapes. It tends to create circular or spherical clusters, even if the true clusters have more complex shapes.

5. **Sensitivity to Outliers:**
   - K-means is sensitive to outliers, and outliers can disproportionately influence the positions of cluster centroids.

6. **Need to Specify \(k\) in Advance:**
   - The number of clusters (\(k\)) needs to be specified before running the algorithm. Selecting an inappropriate \(k\) can lead to suboptimal clustering results.

7. **Does Not Handle Categorical Data:**
   - K-means is designed for numerical data and does not handle categorical features well. Preprocessing steps, such as one-hot encoding, may be required for datasets with categorical variables.

8. **Uniform Cluster Sizes:**
   - K-means tends to produce clusters of roughly equal sizes. In cases where the true clusters have significantly different sizes, this can be a limitation.

9. **Lack of Robustness to Noise:**
   - K-means can be sensitive to noisy data, and outliers can influence the resulting clusters.

10. **May Converge to Local Optima:**
    - The algorithm is susceptible to converging to local optima, especially when the data has a complex structure. Running the algorithm with different initializations can help mitigate this issue.

Despite these limitations, k-means remains a widely used and efficient clustering algorithm, especially in cases where the assumptions align with the characteristics of the data. For scenarios where these assumptions do not hold, alternative clustering methods or preprocessing techniques may be more appropriate.
